```python
import os
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

```python
data_path='/content/drive/MyDrive/Colab Notebooks/cyclegan/apples2oranges'
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
trainA = os.path.join(data_path, 'trainA')
trainB = os.path.join(data_path, 'trainB')
testA = os.path.join(data_path, 'testA')
testB = os.path.join(data_path, 'testB')
```

```python
data_path_resized = '/content/drive/MyDrive/Colab Notebooks/cyclegan/apples2oranges(64*64)'
```

```python
trainA_resized = os.path.join(data_path_resized, 'trainA')
trainB_resized = os.path.join(data_path_resized, 'trainB')
testA_resized = os.path.join(data_path_resized, 'testA')
testB_resized = os.path.join(data_path_resized, 'testB')
```

```python
img_list = os.listdir(trainA)
image = Image.open(os.path.join(trainA, img_list[0]))
```

```python
plt.imshow(image)
```

```python
for img in img_list:
  image = Image.open(os.path.join(trainA, img))
  image = image.resize((64, 64), resample = PIL.Image.BILINEAR)
  image.save(os.path.join(trainA_resized, img))
```

```python
img_list = os.listdir(trainB)
```

```python
for img in img_list:
  image = Image.open(os.path.join(trainB, img))
  image = image.resize((64, 64), resample = PIL.Image.BILINEAR)
  image.save(os.path.join(trainB_resized, img))
```

```python
img_list = os.listdir(testA)
```

```python
for img in img_list:
  image = Image.open(os.path.join(testA, img))
  image = image.resize((64, 64), resample = PIL.Image.BILINEAR)
  image.save(os.path.join(testA_resized, img))
```

```python
img_list = os.listdir(testB)
```

```python
for img in img_list:
  image = Image.open(os.path.join(testB, img))
  image = image.resize((128,128), resample = PIL.Image.BILINEAR)
  image.save(os.path.join(testB_resized, img))
```
